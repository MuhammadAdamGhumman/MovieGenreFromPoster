```python
df_movietotal = pd.read_csv("MovieGenre.csv", encoding='latin1')
print(df_movietotal)
```

```python
df_movietotal_copy = df_movietotal.copy()
df_movietotal_copy = df_movietotal_copy.dropna(axis=0)
```

```python
label_dict = {"word2idx": {}, "idx2word": []}
idx = 0
genre_pm = df_movietotal_copy["Genre"].apply(lambda x: str(x).split("|"))
for l in [g for d in genre_pm for g in d]:
    if l in label_dict["idx2word"]:
        pass
    else:
        label_dict["idx2word"].append(l)
        label_dict["word2idx"][l] = idx
        idx += 1

genre_per_movie = df_movietotal_copy["Genre"].apply(lambda x: str(x).split("|"))

print(genre_per_movie)
```

```python
flat_list = []

def flatten_list(thelist):
    for sublist in thelist:
        for item in sublist:
            flat_list.append(item)
    np.array(flat_list)
    return flat_list
        
flat_list = flatten_list(genre_per_movie)

unique, counts = np.unique(flat_list, return_counts=True)
this = dict(zip(unique, counts))

for key, value in this.items():
    print (key, " = ", value)
```

```python
def preprocess(img, size=(150, 101, 3)):
    img = img.astype(np.float32)
    #img = (img / 127.5) - 1.
    return img
```

```python
images = dict()
for i in range(0, 49):
    images[df_movietotal_copy["imdbId"][i]] = preprocess(cv2.imread(savelocation + str(df_movietotal_copy["imdbId"][i]) + '.jpg'))

#cv2.imwrite('1.jpg', images[df_movietotal_copy["imdbId"][0]])
```

```python
def prepare_data(data, img_dict, label_dict):
    print("Generation dataset...")
    
    dataset = []
    y = []
    ids = []

    for key, value in img_dict.items():
        if key in data["imdbId"].values:
            G = data[data["imdbId"] == key]["Genre"].values
            for g in G: 
                g = g.split("|")

                l = np.sum([np.eye(len(label_dict["idx2word"]) , dtype="uint8")[label_dict["word2idx"][s]] for s in g], axis=0)
                
                y.append(l)
                dataset.append(value)
                ids.append(key)
    print("DONE")
    print(len(dataset))
    return dataset, y, ids

dataset, y, ids =  prepare_data(df_movietotal_copy, images, label_dict)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D, BatchNormalization
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu',
                 input_shape=(268, 182, 3)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(28, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer=keras.optimizers.Adagrad(), metrics=['accuracy'])
```

```python
n = 39
print(len(dataset[: n]))
print(len(y[: n]))
model.fit(np.array(dataset[: n]), np.array(y[: n]), batch_size=3, epochs=3, verbose=1, validation_split=0.1)
```

```python
n_test = 10
X_test = dataset[n:n + n_test]
y_test = y[n:n + n_test]
```

```python
pred = model.predict(np.array(X_test))
```

```python
def accuracy_score(y_test, pred):
    value = 0
    for i in range(0, len(pred)):
        first3_index = np.argsort(pred[i])[-3:]
        correct = np.where(y_test[i] == 1)[0]
        for j in first3_index:
            if j in correct:
                value += 1
    print(value/len(pred))
```

```python
accuracy_score(y_test, pred)
```
